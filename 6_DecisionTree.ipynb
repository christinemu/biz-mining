## 基本概念

### 分类（classification） 

- 给定一条记录$(\boldsymbol{x}, y)$，其中，$y$是分类属性或者目标属性，$\boldsymbol{x}$是该记录预测属性的集合

- 通过学习得到一个**目标函数**（target function）$f$，把每个属性集$\boldsymbol{x}$映射到一个预先定义的**类标签**$y$

$$
y = f(\boldsymbol{x})
$$

- 目标函数也被称作**分类模型**（classification model）

#### 分类模型的功能

1. 描述性建模：识别哪些属性决定一个数据记录属于哪个类别

2. 预测性建模：根据已知的数据记录的属性，自动识别该数据记录属于的类别

#### 分类模型的适用领域

- 非常适合分类属性是**二元**或者**标称**类型的数据集

- 不适用于分类属性是**序数**或**连续**类型的数据集
    + 因为没有考虑标签之间的顺序大小关系

任务|属性集$\boldsymbol{x}$|分类属性$y$
---|---|---
分类e-mail|从e-mail的header和内容中提取的特征|垃圾邮件`or`非垃圾邮件
识别癌变细胞|通过磁共振扫描提取的特征|恶性的`or`良性的
分类星系|从天文望远镜获取的图像中提取特征|椭圆的、螺旋的、`or`不规则星系

#### 分类算法

+ k最近邻分类
+ **决策树**
+ **随机森林**
+ 朴素贝叶斯分类
+ 逻辑回归
+ **神经网络**
+ 支持向量机

#### 混淆矩阵（confusion matrix）

- 由分类模型做出的正确和错误的分类结果构成的矩阵

- 二元分类问题的混淆矩阵

![](./img/classification/confusionMatrix.gv.svg)

#### 性能度量（performance metric）

- 汇总混淆矩阵的信息

- 准确率（accuracy）

$$
准确率=\frac{正确预测数}{预测总数}=\frac{f_{11}+f_{00}}{f_{11}+f_{10}+f_{01}+f_{00}}
$$

- 错误率（error rate）

$$
错误率=\frac{错误预测数}{预测总数}=\frac{f_{10}+f_{01}}{f_{11}+f_{10}+f_{01}+f_{00}}
$$

$$
错误率+准确率=1
$$

- 最佳选择划分通常根据结点的**不纯性的程度**（degree of impurity）

- 不纯的程度越高，类分布就越倾斜

#### 不纯性度量 

令$p(i|t)$表示给定结点$t$中属于类$i$的记录所占的比例，类别个数为$c$

- 熵（Entropy）：衡量系统中混乱或不确定性的程度

$$ \text{Entropy}(t)=-\sum_{i=0}^{c-1}p(i|t)\log_2p(i|t) $$

其中，$0\times\log_2 0=0$

> ID3、C4.5算法用熵选择最佳划分

- Gini系数

$$ \text{Gini}(t)=1-\sum_{i=0}^{c-1}\left[p(i|t)\right]^2 $$

- Gini系数可以直观认为是衡量分类错误的概率

$$\text{Gini}(t)=\sum_{i=0}^{c-1}p(i|t)\bigl(1-p(i|t)\bigr)$$

> CART算法用Gini系数选择最佳划分

- 分类错误（classification error）

$$
Ce(t) = 1-\underset{i}{\max}\bigl(p(i|t)\bigr)
$$

假设一个节点中只包含有两个类别（类别1和类别2），各自所占的比例是$p_1$和$p_2$，且$p_1+p_2=1$

````python
x = np.linspace(0.01,0.99)
x1 = np.ones(x.shape) - x
xE = -x*np.log2(x)-x1*np.log2(x1)
xG = 1-(np.power(x,2)+np.power(x1,2))
XCe = [1-max(z,1-z) for z in x]
df = pd.DataFrame({'x':x,'Entropy':xE,'Gini':xG,'Classification error':XCe})
ax = df.plot(x='x',y='Entropy',kind='line',figsize=(12,6))
df.plot(x='x',y='Gini',kind='line',ax=ax)
df.plot(x='x',y='Classification error',kind='line',ax=ax)
ax.set(xlabel='$p_1$')
````

#### 确定测试条件

- 对于选择的一个测试条件（特征），计算父结点（划分前）的不纯度与子结点（划分后）的不纯度的差(即信息增益，information gain)，差越大，测试条件的效果就越好

对一个内部节点$p$，选择特征$f$划分的信息增益的计算如下：

$$IG(D_p, f)=I(D_p)-\sum_{j=1}^c\frac{N_j}{N_p}I(D_j)$$

其中，$D_p$和$D_j$分别是父节点和划分后的子节点对应的数据集，$I(\cdot)是不纯度的度量$，$N_p$和$N_j$分别是父节点和划分后的子节点中的样本数量

$$\underset{f\in\mathcal{F}}{\max}IG(D_p,f)$$

其中，$\mathcal{F}$是特征的集合

- 通常为了降低搜索空间，决策树算法通常构建二叉树，即，每个父节点只被划分成**两个**子节点。此时的信息增益为

$$IG(D_p,f)=I(D_p)-\frac{N_\text{left}}{N_p}I(D_\text{left})-\frac{N_\text{right}}{N_p}I(D_\text{right})$$

#### 分类性能度量 

##### 混淆矩阵
- 由分类模型做出的正确和错误的分类结果构成的矩阵

<center><img src="./img/classification/PosNeg.svg" width=60%></center>

- **准确率（accuracy score）**

$$
准确率=\frac{正确预测数}{预测总数}=\frac{TP+TN}{TP+FP+FN+TN}
$$

- **精确率（precision score, 查准率）**
$$
precision = \begin{cases}
\frac{TP}{TP+FP},\quad \text{类别}P\\
\frac{TN}{TN+FN},\quad \text{类别}N
\end{cases}$$

- 针对某个类别的样本，不做出错误分类的能力

- **召回率（recall score，查全率）**
$$recall = \begin{cases}
\frac{TP}{TP+FN},\quad \text{类别}P\\
\frac{TN}{TN+FP},\quad \text{类别}N
\end{cases}$$

- 能够识别所有属于某个真实类别的样本的能力

- `estimator`：训练好的分类器

#### 先剪枝（Forward-Pruning）

<p class="alert alert-info">提前停止树的构造而对树进行剪枝</p>

- 在决策树到达一定高度的情况下就停止树的生长
- 到达结点的样本个数小于某一个阈值可停止树的生长

```python
tree.DecisionTreeClassifier(max_depth=None, min_samples_split=2, min_samples_leaf=1)
```

- `max_depth`：`int`类型或`None`，树的最大深度。若为`None`，则所有的叶结点都只包含纯类，或者所有叶结点包含的样本数量小于`min_samples_leaf`
    + 值过大会导致算法对训练集的**过拟合**，而过小会妨碍算法对数据的学习
    + 推荐初始设置为3，先观察生成的决策树对数据的初步拟合状况，再决定是否要增加深度

- `min_samples_split`：`int`类型或`float`类型，划分一个内部结点需要的最少的样本数量。
    + `int`类型，`min_samples_split`为最小值，默认是2个样本
    + `float`类型，在全部样本中的占比，`ceil(min_samples_split * n_samples)`为最小值
    + 值越大，决策树的枝越少，达到一定的先剪枝效果

- `min_samples_leaf`：`int`类型或`float`类型，每个叶结点需要包含的最少的样本数量。
    + `int`类型，`min_samples_leaf`为最小值，默认是1个样本
    + `float`类型，在全部样本中的占比，`ceil(min_samples_leaf * n_samples)`为最小值
    + 值越大，决策树的枝越少，达到一定的先剪枝效果

#### 后剪枝（Post-Pruning）

<p class="alert alert-danger">构造完整的决策树，然后用叶结点替换那些置信度不够的结点的子树，该叶结点所应标记的类别为被替换的子树中大多数样本所属的类别。</p>

- 利用成本复杂度剪枝（cost-complexity pruning）

对于决策树$T$，定义成本复杂度的测量函数$R_{\alpha}(T)$，

$$R_{\alpha}(T)=R(T)+\alpha|\tilde{T}|$$

其中，$R(T)$是所有叶结点总不纯度（由叶结点包括的样本数量加权的）；$|\tilde{T}|$是决策树$T$中所有叶结点的数量。$\alpha$是**复杂度系数**。

- 如何计算$\alpha$？

    - 定义子树$T_t$为决策树$T$中的一个内部结点$t$与该内部结点的所有子结点。
    - 一般，$R(T_t)<R(t)$
    - $\alpha$的取值使得该内部结点本身的成本复杂度`=`子树$T_t$的成本复杂度，即

$$R_{\alpha}(T_t)=R_{\alpha}(t)$$

$$
\alpha = \frac{R(t)-R(T_t))}{|T|-1}
$$

- 一个拥有较小$\alpha$的内部节点将会被剪枝

```python
tree.DecisionTreeClassifier(ccp_alpha=None)
```
- `ccp_alpha`：非负的浮点数，子树的复杂度系数，复杂度系数低于该参数的子树会被剪枝

\begin{problem}\label{prob:boostrap}
每个boostraping的样本平均包含多少个原始数据集中的数据？
\end{problem}


\begin{example}\label{example:boostrap}
假设有一个数据集含有$n$个样本点，但是生成数据的分布未知。估计该数据集的中位数来估计分布本身的中位数，如何计算该估计的置信度？
\end{example}


- 通过有放回的随机抽样从原始数据集中生成含有$n$个数据点的新的数据集

````python
def bootstrap_sample(data):
    """有放回的随机抽样len(data)个数据"""
    return np.random.choice(data,len(data),replace=True)
````

```python
import numpy as np
np.random.choice(a, size=None, replace=True, p=None)
```

- `a`：1维序列或者`int`，原始数据集。如果是`int`，则从`np.arange(a)`中抽取数据
- `size`：`int`或者由`int`组成的元组。输出数据的形状，如果输出形状为`(m,n,k)`，则抽样的数据数量为`m*n*k`。默认只抽取一个数据
- `replace`：`bool`变量，是否有放回抽样
- `p`：1维序列，`a`中每个样本的权重。默认每个样本等权重
- 返回抽样得到的样本

````python
def bootstrap_statistic(data, stats_fn, num_samples):
    """从data中bootstrap抽样num_samples个样本，在这些样本上进行统计计算stats_fn"""
    return [stats_fn(bootstrap_sample(data)) for _ in range(num_samples)]
````

#### Bagging原理

- 原始数据集是$D$
- 采用Bootstraping方法生成数据集$D_i,i=1,2,\cdots,K$
- 在数据集$D_i$上训练分类器$M_i$

- 给定一个检验数据点$\mathbf{x}$
- 用$K$个分类器逐一预测$\mathbf{x}$对应的类别。将$\mathbf{x}$预测为$c_j$类的分类器的数量记为$v_j(\mathbf{x})$，

$$
v_j(\mathbf{x})=\left|\{M_i(\mathbf{x})=c_j | i=1,2,\cdots,K\}\right|
$$

- 集成的分类器，记为$\mathbf{M}^K$，通过**投票**（majority voting）方式预测$\mathbf{x}$最终所属的类别

$$\mathbf{M}^K({\mathbf{x}})=\underset{c_j}{\arg \max}\{v_j(\mathbf{x})|j=1,2,\cdots,k\}$$

#### 集成分类器误差

- 假设有$n$个基础分类器，拥有相同的错误率$\varepsilon$
- 假设基础分类器相互独立，即错误率相互独立
- 集成分类器的错误率为

$$\sum_k^nC_n^k\varepsilon^k(1-\varepsilon)^{n-k}$$

````python
from scipy.special import comb
import math
def ensemble_error(n_classifier,error):
    """给定n_classifier个基础分类器的错误率error，计算集成分类器的错误率"""
    k_start = math.ceil(n_classifier/2.)
    probs = [comb(n_classifier,k)*error**k*(1-error)**(n_classifier-k) for k in range(k_start,n_classifier+1)]
    return sum(probs)
````

````python
ensemble_error(11,0.25)
````

````python
# 绘制集成分类器错误率随基础分类器错误率的变动趋势

base_errors = np.arange(0.0,1.01,0.01)
ens_errors = [ensemble_error(11,error) for error in base_errors]
error_df = pd.DataFrame({'base':base_errors,'ens':ens_errors})
error_df.plot(kind='line',style=['-','--'])
````

- 只要基础分类器的错误率**低于0.5**，集成分类器的错误率就会**低于**单一基础分类器

### 随机森林

#### 随机森林基本原理

- 组合多个决策树，通过投票的方式确定样本所属的类别，使得整体模型在抑止过度拟合的同时，获得较好的精确度

1. 从训练集中bootstrap抽样$n$个样本
1. 从bootstrap样本中生成决策树。在每个节点
    - 用**无放回**的方法随机抽样$d$个属性
    - 依据最大化信息增益选择属性实现最优划分
1. 重复1、2两个步骤$k$次（共建立$k$个决策树）
1. 通过投票方式集成所有决策树的预测

- 随机森林的两个随机源
    - bootstrap样本
    - 随机选择属性

<center><img src="./img/classification/randomForest.jpg" width=100%></center>

#### 影响随机森林分类效果的参数

- 决策树数量
    - 一般而言决策树的数量越多，随机森林的分类效果越好

- Bootstrap抽样的大小$n$

### Boosting

#### 基本原理

<center><img src="./img/classification/AdaBoost.png" width=70%></center>

- 在不同的样本上训练基础分类器

- 挑选样本以提升分类器在难以区分样本上的准确性
    - 上一轮中分类错误的样本在下一轮训练中会给予更高的权重

- 最终通过加权投票（weighted voting）的方式得到最终结果

#### Adaptive Boosting (AdaBoost) 

- 假设:
    - $\mathbf{D}$是训练集，包含样本点$\vec{x}_i\in\mathbb{R}^d, i=1,2,\cdots,n$
    - 总共训练$K$次，每次训练用$t=1,2,\cdots,K$表示
    - 令$\alpha_t$表示第$t$次训练得到的第$t$个分类器$M_t$的权重
    - 令$w_i^t$表示样本$\vec{x}_i$在第$t$次训练中的权重，$\vec{w}^0=\frac{1}{n}\vec{1}$

- 在第$t$次迭代中，通过$\vec{w}^{t-1}$决定的分布生成训练样本集$\mathbf{D}_t$
- 用$\mathbf{D}_t$训练分类器$M_t$，并计算该分类器在训练集$\mathbf{D}$上的加权错误率
$$
\epsilon_t=\sum_{i=1}^nw_i^{t-1} \cdot I\bigl(M_t(\vec{x}_i)\ne y_i\bigr)
$$

- 第$t$轮训练得到的分类器的权重为

$$
\alpha_t=\ln\bigl(\frac{1-\epsilon_t}{\epsilon_t}\bigr)
$$

- 根据每个样本点$\vec{x}_i\in\mathbf{D}$是否被分错，调整它的权重
$$
w_i^t=w_i^{t-1}\cdot\exp\{\alpha_t\cdot I(M(\vec{x}_i)\ne y_i))\}
$$

- 若未分错，则
$$
w_i^t=w_i^{t-1}
$$

- 若分错，则
$$
w_i^t=w_i^{t-1}\cdot \exp\{\alpha_t\}=w_i^{t-1}\bigl(\frac{1}{\epsilon_t}-1\bigr)
$$

- 规范化权重向量

$$
\vec{w}^t=\frac{\vec{w}^t}{\vec{1}^T\vec{w}^t}
$$

- 给定顺次训练好的分类器$M_t, t=1,2,\cdots,K$和对应的权重$\alpha_t$，一个检验样本$\vec{x}$的类别由加权投票决定

$$
v_j(\vec{x})=\sum_{t=1}^K\alpha_t\cdot I\bigl(M_t(\vec{x})=c_j\bigr)
$$

其中，$v_j(\vec{x})$为类别$c_j$的加权投票

合成的分类器$\mathbf{M}^K$预测$\vec{x}$的类别为

$$
\mathbf{M}^K(\vec{x})=\underset{c_j}{\arg\max}\{v_j(\vec{x})\lvert j=1,2,\cdots,k\}
$$

```python

```

- `estimator`：基础分类器，默认为Decision Tree Stump (最大深度为1)
